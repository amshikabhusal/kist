```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
rainfall = [14,18,35,62,145,300,375,340,220,80,12,5]
df = pd.DataFrame({
'Month': months,
'Rainfall': rainfall
})
print(df)
```

```python
print("Average Rainfall:", np.mean(rainfall))
print("Maximum Rainfall:", np.max(rainfall))
print("Minimum Rainfall:", np.min(rainfall))
```

```python
plt.plot(df['Month'], df['Rainfall'], marker='o')
plt.title("Monthly Rainfall in Nepal (Kathmandu)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Month'], df['Rainfall'], color='blue')
plt.title("Monthly Rainfall Comparison")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
plt.pie(df['Rainfall'], labels=df['Month'], autopct='%1.1f%%')
plt.title("Rainfall Distribution")
plt.show()
```

```python
plt.hist(df['Rainfall'], bins=5)

plt.title("Rainfall Distribution Histogram")
plt.xlabel("Rainfall")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['Rainfall'])
plt.title("Rainfall Box Plot")
plt.show()
```
